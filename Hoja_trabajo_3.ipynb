```python
class regresion():
    ##Declaro variables
    xlist = [] 
    ylist = []
    wlist = []
    beta0 = 0
    beta1 = 0
    x1 = 0
    x2 = 0
    r2 = 0
    r = 0
    n = 0   
    xy = 0
    sumax = 0
    sumay = 0
    sumaxy=0
    y1 = 0
    y2 = 0
    n = 0
    cuadradox = 0
    cuadradoy = 0
    cuadradosumx = 0
    cuadradosumy = 0
    
    def __init__(self, x, y):
        ## X Y Y LISTAS DE NUMEROS DE LA MISMA CANTIDAD DE ELEMENTOS    
        self.xlist = x
        self.ylist = y

        if len(self.xlist)== len(self.ylist):
            #Listo variables para regresion
            self.n = len(self.xlist)
            i = 0 #iterador 
            while i < self.n:
                self.xy= float(self.xlist[i]) * float(self.ylist[i])
                self.sumaxy += self.xy
                self.sumax += float(self.xlist[i])
                self.sumay += float(self.ylist[i])
                self.cuadradosumx += pow(float(self.xlist[i]),2)
                self.cuadradosumy += pow(float(self.ylist[i]),2)
                i+=1
            self.x1 = self.sumax / self.n
            self.y2 = self.sumay / self.n
            self.sumay = self.sumay
            self.cuadradox = pow(self.sumax,2)
            self.cuadradoy = pow(self.sumay,2)
            self.beta0= ((self.n * self.sumaxy) - (self.sumax * self.sumay)) / ((self.n * self.cuadradosumx) - (self.cuadradox))
            self.beta1 =((self.sumax * self.sumaxy) - (self.sumay * self.cuadradosumx)) / (self.cuadradox - (self.n * self.cuadradosumx))
            self.listaResultado=[]
            self.listaResultado.append(self.beta0)
            self.listaResultado.append(self.beta1)
            
            self.r2()
            self.r()
            self.getAllParams()
            
            sPredecir = input('\nDesea predecir uno o varios valores? S / N')
            
            if (str(sPredecir).upper() == 'S'):
                tipo = int(input('\nIngrese el tipo de prediccion:\n 1 - Para prediccion de un solo valor\n 2 - Para prediccion de varios valores'))
                
                if(tipo == 1):
                    prediccion = input('\n\n Ingrese un valor para la predicciÃ³n:')
                    prediccion = float(prediccion)
                    self.predict(prediccion)
                else:
                    prediccion = int(input("\n\nIngrese la cantidad de datos a ingresar a la lista"))
                    contador = 1
                    listax = []
                    while contador <= prediccion:
                        nuevo = float(input("Ingrese elemento:"))
                        listax.append(nuevo)
                        contador += 1
                    self.predictN(listax)
                print('Fin de la ejecucion')
            else:
                print('Fin de la ejecucion')
            
        else:
            print('Las listas deben coincidir en elementos')
    
    def predict(self,val):
        self.y = self.beta0 + (self.beta1 * val)
        print("Y = b0 + b1x =" + str(self.y) )      
    
    def predictN(self,vals):
        self.wlist = vals
        
        i=0 #iterador
        while i < len(self.wlist):
            self.y= (self.beta0 + (self.beta1 * self.wlist[i]))
            print("Y"+ str(i) +"= b0 + b1x =" + str(self.y) )    
            i+= 1            
    
    def r2(self):
        self.r2 = ((self.n * self.sumaxy)-(self.sumax * self.sumay))/((((self.n * self.cuadradosumx ) - self.cuadradox ) * ( ( self.n * self.cuadradosumy ) - self.cuadradoy))**2)
        return(self.r2)
        
    def r(self):
        self.r = (self.r2**2)
        return(self.r)
       
    def getAllParams(self):
        print('{b0 : '+ str(self.beta0) +', b1 : '+ str(self.beta1) +', r2 : '+ str(self.r2) +', r:'+ str(self.r) +'}')
        
    @classmethod
    def from_file(cls,path):
        archivo = open(path,"r")  
        
        lista=[]
        for lineas in archivo:
            lineas = lineas.rstrip()
            lista.append(lineas.split(','))
        lista.pop(0)
        archivo.close()
        
        x=[]
        y=[]
        #Recorro lista de valores
        for  item, subLista in enumerate(lista):
            #Recorro sublista y divido en X y Y
            for subItem in enumerate(subLista):
                if subItem[0]==0:
                    x.append(subItem[1])                    
                else:
                    y.append(subItem[1])
        
        regresion(x,y)
     
```
