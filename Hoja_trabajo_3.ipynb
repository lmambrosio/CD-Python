```python
class regresion():
    def __init__(self, x, y):
        ## X Y Y LISTAS DE NUMEROS DE LA MISMA CANTIDAD DE ELEMENTOS    
        
        
        self.x = x
        self.y = y
        self.beta0 = 0
        self.beta1 = 0
        self.x1 =0
        self.x2 =0
        self.sumay=0
        self.r2=0
        self.r=0
        
        if len(self.x)== len(self.y):
            #Listo variables para regresion
            n = len(self.x)
            xy = 0
            sumax = 0
            sumay = 0
            sumaxy=0
            y1 = 0
            y2 = 0
            cuadradox = 0
            cuadradosumx = 0
            
            i = 0 #iterador 
            while i < n:
                xy= float(self.x[i]) * float(self.y[i])
                sumaxy+=xy
                sumax += float(self.x[i])
                sumay += float(self.y[i])
                cuadradosumx += pow(float(self.x[i]),2)
                i+=1
            
            self.x1 = sumax / n
            self.y2 = sumay / n
            self.sumay = sumay
            cuadradox = pow(sumax,2)
            
            self.beta0= ((n * sumaxy) - (sumax * sumay)) / ((n * cuadradosumx) - (cuadradox))
            self.beta1 =((sumax * sumaxy) - (sumay * cuadradosumx)) / (cuadradox - (n * cuadradosumx))
            
            self.listaResultado=[]
            self.listaResultado.append(self.beta0)
            self.listaResultado.append(self.beta1)
            #print(self.beta0)
            #print(self.beta1)                     
        else:
            print('Las listas deben coincidir en elementos')
        
    
   ### def predict(self,val):
    ###Este m´etodo devuelve la predicci´on de la regresi´on aplicada sobre el valor que recibe como par´ametro
   ###     #codigo
   ### 
   ### def predictN(self,vals):
    ### Este m´etodo devuelve la predicci´on de la regresi´on aplicada sobre una lista de valores que recibe como par´ametro.
   ###     #codigo
   ### 
    def r2(self):
        r2 = pow(self.x1-self.y2,2) / pow(self.x1-self.sumay,2)
        return(r2)
    def r(self):
        self.r = (self.r2**2)
       
    def getAllParams(self):
        #  +', b1 : '+ str(self.beta1) +', r2 : '+ str(self.r2) +', r:'+ str(self.r) +'}'
       ## return self.beta0
        lista=[]
        for item in enumerate(self.listaResultado):
            lista.append(item)
        
        return lista
   ###     #codigo {b0 : valor, b1 : valor, r2 : valor, r : valor}
        
    @classmethod
    def from_file(cls,path):
        archivo = open(path,"r")  
        
        lista=[]
        for lineas in archivo:
            lineas = lineas.rstrip()
            lista.append(lineas.split(','))
            ##print(lineas.split(','))
        lista.pop(0)
        archivo.close()
        
        x=[]
        y=[]
        #Recorro lista de valores
        for  item, subLista in enumerate(lista):
            #Recorro sublista y divido en X y Y
            for subItem in enumerate(subLista):
                if subItem[0]==0:
                    y.append(subItem[1])                    
                else:
                    x.append(subItem[1])
        
        regresion(x,y) 
        regresion.r2()
```
